# My first scientific app in python with numpy
In this first example we will create a sine function and plot it. The equation is: $y=3sin(2\pi x)$, where $x \in[-1, 1]$.

(1) Let's import the numpy library, and define cariables x and y

```python
import numpy as np
# create the values in x. This is a vector (list)
x = np.linspace(-1, 1, 1000)
y = 3*np.sin(2*np.pi*x)
```

(2) Import the graphic library and show the function y

```python
import matplotlib.pyplot as plt
%matplotlib inline
# this is needed to show the graphs in the notebook
plt.plot(x, y)
plt.show
```

(3) Calculate the differential of $y$ and show it

```python
dy_dx = np.diff(y)
plt.plot(x[:-1], dy_dx)
```

```python
# (3) Calculate the differential of $y$ and show it.
dy_dx = np.diff(y)
plt.plot(x[:-1], dy_dx)
plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$x$', fontsize = 14)
plt.title('$y=cos(2\pi x)$', fontsize = 14)


```

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2*np.pi*t)
plt.plot(t,s)

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.grid(True)
plt.savefig('test.png')
plt.show()
```

## Fun with statistics: app No 2
Lets create a random sample of 1000 points from a Poisson distribution

```python
p = np.random.poisson(5, 1000)
plt.plot(p, color='r')
plt.grid()
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
```

```python
mu = np.mean(p,axis=0)
print('The expected value (mean) of the normal distribution is {}'.format(mu))
```

```python
b,c,d = plt.hist(p, 15, normed=True)
```

```python
print('The total probability is {}'.format(sum(b)))
```

```python
cdf = np.cumsum(b)
plt.plot(cdf, color='black', linewidth=3)
plt.xlim([0,10])
plt.xlabel('Random number $x$', fontsize=14)
plt.ylabel('Cumulative distribution', fontsize=14)
```
