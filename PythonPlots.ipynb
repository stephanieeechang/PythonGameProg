# Working with subplots

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# creates an array of nimbers from 0 to 1 with 1000 points
x = np.linspace(0, 1, 1000)
y = np.cos(x**2*np.pi*15)

plt.plot(x,y)
```

```python
# using subplots
f, axis = plt.subplots()
axis.plot(x, y, color='m')
axis.set_title('$y=cos(15\pi x^2)$')
```

```python
# defines container of the graph
f, axis2 = plt.subplots(2)
axis2[0].plot(x,y)
axis2[1].plot(x,[1/i for i in y], color='r')
```

```python
f, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x,y)
ax1.set_title('sharing y axis')
ax2.plot(x,[y[i] for i in range(len(y)-1, -1, -1)], color='b')
ax2.set_title('reversed y')
```

```python
f, (a1, a2) = plt.subplots(2, 2)
a1[0].plot(x, y)
a1[0].set_title('sharing y axis')
a1[1].plot(x,[y[i] for i in range(len(y)-1, -1, -1)], color='g')
a1[1].set_title('reversed y')
a2[0].plot(x, np.sqrt(y+1), color='y')
a2[0].set_title('square root of y')
a2[1].plot(x, np.abs(y), color='r')
a2[1].set_title('absolute value of y')
```

# population of students a PAS

```python
age = np.random.normal(10, 3, 500)
height = np.random.normal(150, 10, 500)
# plt.hist(age)
# plt.hist(height)

fig = plt.figure()
ax = fig.add_subplot(2, 2, 1)
plt.scatter(age, height, color='r')

ax = fig.add_subplot(2, 2, 2)
ax.hist(height, color='r', orientation='horizontal')

ax = fig.add_subplot(2, 2, 3)
ax.hist(age, color='r')
```

# Pie Chart in Python

```python
pets = ['dogs', 'cats', 'horses', 'sea horses', 'unicorns']
freq = [10, 2, 20, 10, 55]

fig1, ax1 = plt.subplots() # container

ax1.pie([f/sum(freq) for f in freq], labels=pets, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
```
