# Working with images in Python

```python
from skimage import data # library to work with Images SkImages
import numpy as np

cat = data.chelsea()
print('The size of the image is {}'.format(np.shape(cat)))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(cat)
```

```python
redComp = cat[:,:,0]
# cmap=summer (green to yellow)
plt.imshow(redComp)
```

```python
# using a different color representation
plt.imshow(redComp, cmap='binary')
plt.colorbar()
```

```python
import matplotlib.image as img
prism = img.imread('RGB_illumination.jpg')
# plt.imshow(prism)
prismRed = prism[:,:,0]
plt.imshow(prismRed, cmap='binary')
plt.colorbar()
```

```python
# histogram of the image
n, b, p = plt.hist(prismRed.ravel(), bins=100, range=(0, 255), normed=True)
```

```python
from skimage import exposure
prismRedEq = exposure.equalize_hist(prismRed)
n, b, p = plt.hist(prismRed.ravel(), bins=100, range=(0, 1), normed=True)
```
